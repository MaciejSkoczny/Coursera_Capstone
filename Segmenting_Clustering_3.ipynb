```python
import folium
import requests 
import json 
import matplotlib.cm as cm
import matplotlib.colors as colors
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from pandas.io.json import json_normalize 
from sklearn.cluster import KMeans
from geopy.geocoders import Nominatim 
```

Reading the file created in Part 2

```python
df = pd.read_csv('tor2.csv')
df
```

Creating a data frame that includes only Toronto boroughs

```python
map_df = df[df['Borough'].str.contains('Toronto')].reset_index(drop=True)
map_df.head(10)
```

Creating a map of Toronto based on the data from the new data frame (starting with M4E)

```python
map_tor = folium.Map(location=[df["Latitude"][0], df["Longitude"][0]], zoom_start=10)
map_tor
```

```python
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='violet',
        fill=True,
        fill_color='#1515cc',
        fill_opacity=0.5,
        parse_html=False).add_to(map_tor)  
    
map_tor
```
